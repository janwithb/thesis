```python
import gym
import os
import dmc_remastered as dmcr
import matplotlib.pyplot as plt
import numpy as np
import envs

from dm_control import suite
from wrappers.gym_wrapper import GymWrapper

os.environ['KMP_DUPLICATE_LIB_OK']='True'
```

```python
images = []
envs = ['cartpole', 'reacher', 'cheetah', 'walker']
tasks = ['swingup', 'easy', 'run', 'walk']
for env, task in zip(envs, tasks):
    env = suite.load(env, task, task_kwargs={'random': 0})
    env = GymWrapper(env)
    env.reset()
    images.append(env.render(mode='rgb_array', height=256, width=256, camera_id=0))
    env.close()
    
for env, task in zip(envs, tasks):
    _, env = dmcr.benchmarks.visual_generalization(env, task, num_levels=100)
    env.reset()
    images.append(env.render(mode='rgb_array', height=256, width=256, camera_id=0))
    env.close()
    
# create figure
fig = plt.figure(figsize=(24, 12))
columns = 4
rows = 2

for i in range(len(images)):
    fig.add_subplot(rows, columns, i + 1)
    plt.imshow(images[i])
```
